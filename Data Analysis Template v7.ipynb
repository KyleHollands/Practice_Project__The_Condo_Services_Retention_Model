#### **Task 1b. Load Data**

```python
# Generate the dataset
np.random.seed(42)

n_buildings = 1500

# 1. Building Characteristics
data = {
    'building_id': range(1001, 1001 + n_buildings),
    'region': np.random.choice(['North', 'South', 'East', 'West'], n_buildings, p=[0.6, 0.1, 0.1, 0.2]),
    'condo_size_units': np.random.randint(10, 500, n_buildings), # 10 units to 500 units
    'contract_type': np.random.choice(['Annual_Subscription', 'Monthly_Retainer', 'Transactional'], n_buildings, p=[0.4, 0.3, 0.3]),
    'account_age_years': np.random.randint(1, 15, n_buildings),
}

df = pd.DataFrame(data)

# 2. Services Used (The "Dirty" String Column)
# We need to generate comma-separated strings like "Cleaning,Security"
service_options = ['Cleaning', 'Security', 'Concierge', 'Landscaping', 'Pool_Maint']

def assign_services(row):
    # Larger buildings use more services
    n_services = np.random.randint(1, 4)
    if row['condo_size_units'] > 200:
        n_services += 1
    
    selected = np.random.choice(service_options, size=min(n_services, 5), replace=False)
    return ",".join(selected)

df['services_used'] = df.apply(assign_services, axis=1)

# 3. Financials
# Spend correlates with size + randomness
df['annual_spend'] = (df['condo_size_units'] * np.random.uniform(500, 1500)) 
# Transactional clients spend less/irregularly
df.loc[df['contract_type'] == 'Transactional', 'annual_spend'] *= 0.4

# 4. Generate Churn Target
def calculate_churn_risk(row):
    score = 0.10 # Base risk
    
    # RISK DRIVER 1: Price Sensitivity (Small building, High Spend)
    spend_per_unit = row['annual_spend'] / row['condo_size_units']
    if spend_per_unit > 1200:
        score += 0.30
        
    # RISK DRIVER 2: Contract Type
    if row['contract_type'] == 'Transactional':
        score += 0.25 # No loyalty
    elif row['contract_type'] == 'Annual_Subscription':
        score -= 0.10 # Locked in
        
    # RISK DRIVER 3: "Sticky" Services
    # If they use 'Concierge', they are harder to churn (human relationship)
    if 'Concierge' in row['services_used']:
        score -= 0.15
        
    # RISK DRIVER 4: Legacy Clients
    if row['account_age_years'] > 10:
        score -= 0.10
        
    # Add noise
    score += np.random.normal(0, 0.1)
    
    return min(max(score, 0), 1)

df['churn_prob'] = df.apply(calculate_churn_risk, axis=1)
df['churn'] = df['churn_prob'].apply(lambda x: 1 if x > 0.5 else 0)

# Filter dataset to match your business case (Focus on 'North' Region)
# But keep the column so you can filter it IN your template, not before.
print("Dataset Generated!")
print(df.head())
```

##### **Statistical Testing Metrics Reference** 📊

- **P-value:** Probability of observing results as extreme as those measured, assuming the null hypothesis is true (commonly, < 0.05 indicates statistical significance).
- **Z-statistic:** Standardized value showing how many standard deviations a sample mean is from the population mean; used for large samples or when population variance is known.
- **T-statistic:** Measures the difference between group means relative to sample variability; used for smaller samples or when population variance is unknown.
- **Effect Size (Cohen's d):** Quantifies the magnitude of the difference between groups, regardless of sample size.
  - Small: 0.2, Medium: 0.5, Large: 0.8
- **Confidence Interval:** Range of values within which the true population parameter is likely to fall, with a given level of confidence (e.g., 95%).
- **Statistical Power:** Probability that a test will detect a true effect when it exists (aim for 80%+ to reduce risk of Type II error).

*Tip: Always report both statistical significance (p-value) and practical significance (effect size/confidence interval) for context.*

##### **Check Statistical Significance of Features for Modeling**

```python
# Check statistical significance of the features with respect to the target variable (e.g., using t-tests for continuous features and chi-squared tests for categorical features).
def statistical_significance_tests(df, target_col):
    from scipy.stats import ttest_ind, chi2_contingency
    
    print("🔬 Statistical Significance Tests")
    print("="*60)
    print(f"Significance level (α): 0.05\n")
    
    significant_features = []
    
    # Exclude non-predictive columns
    columns_to_skip = [target_col, 'building_id', 'region', 'services_used']
    
    for column in df.columns:
        if column in columns_to_skip:
            continue

        # Numeric feature -> T-test
        if pd.api.types.is_numeric_dtype(df[column]):
            group1 = pd.to_numeric(df.loc[df[target_col] == 0, column], errors='coerce').dropna()
            group2 = pd.to_numeric(df.loc[df[target_col] == 1, column], errors='coerce').dropna()

            if len(group1) > 1 and len(group2) > 1:
                t_stat, p = ttest_ind(group1, group2, equal_var=False, nan_policy='omit')
                significance = "✅ SIGNIFICANT" if p < 0.05 else "❌ Not significant"
                print(f"{column:25s} | p-value = {p:.4f} | {significance}")
                if p < 0.05:
                    significant_features.append(column)
        else:
            # Categorical feature -> Chi-squared test
            contingency_table = pd.crosstab(df[column].astype(str), df[target_col])
            if contingency_table.shape[0] > 1 and contingency_table.shape[1] > 1:
                chi2, p, dof, expected = chi2_contingency(contingency_table)
                significance = "✅ SIGNIFICANT" if p < 0.05 else "❌ Not significant"
                print(f"{column:25s} | p-value = {p:.4f} | {significance}")
                if p < 0.05:
                    significant_features.append(column)
    
    print(f"\n📊 Summary: {len(significant_features)} significant features identified")
    print(f"Significant features: {significant_features}")
    return significant_features

# Run tests on RAW features only
significant_features = statistical_significance_tests(df, 'churn')
```

#### **Task 4b. Handle Outliers**

Outliers can have varying impacts depending on the modeling approach. Always visualize and assess outliers before modeling. Consider robust methods or transformations if outliers are present.

| Model               | Importance of Handling Outliers | Why?                                                                                                                                                                                                 |
|---------------------|---------------------------------|----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------|
| Decision Trees (DT) | Low                             | Decision Trees split data based on thresholds, making them inherently robust to outliers. Extreme outliers may slightly influence split thresholds, especially in shallow trees.                    |
| Random Forests (RF) | Low to Moderate                 | Random Forests average predictions across multiple trees, reducing the influence of outliers. Consistent outliers across samples may slightly affect feature importance rankings.                   |
| Linear Regression   | High                            | Outliers can disproportionately influence the regression line, distorting coefficients and residuals, leading to poor model performance and interpretability.                                      |
| Logistic Regression | Moderate to High                | Outliers can affect the decision boundary, especially if they dominate the feature space. The sigmoid function reduces sensitivity, but extreme outliers can still skew predictions.               |
| XGBoost             | Moderate                        | XGBoost is generally robust to outliers due to its tree-based nature and regularization. However, dominant outliers can influence split thresholds and feature importance, especially early on.     |
| Statistical Testing | High                            | Outliers can strongly affect tests (t-tests, ANOVA, correlation) by inflating variance and distorting means. Non-parametric tests (e.g., Mann-Whitney U, Wilcoxon) are less sensitive. Always check for outliers and consider robust alternatives if needed. |

**Tip:**  
- Use boxplots or histograms to visually inspect outliers.
- Consider winsorizing, transformation, or robust models if outliers are problematic.

##### **Check for Outliers**

```python
# Check for outliers in numeric columns (excluding target variable 'churn') using IQR
def detect_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    
numeric_columns = [col for col in df.select_dtypes(include=[np.number]).columns if col != 'churn']

for column in numeric_columns:
    outliers = detect_outliers_iqr(df, column)
    print(f"Number of outliers in {column}: {outliers.shape[0]}")
```

#### **Task 4h. Remove Highly Correlated Features (VIF)**

```python
# Iterative VIF-based removal with protected features
print("\nIterative VIF-based feature removal...")

# PROTECTED FEATURES: Won't be removed regardless of VIF
protected_features = ['spend_per_unit']  # Engineered features to keep
print(f"Protected features (will not be removed): {protected_features}")

X = df.drop(columns=['churn'])
max_vif_threshold = 10

while True:
    vif_data = pd.DataFrame()
    vif_data["feature"] = X.columns
    
    vif_values = []
    for i in range(X.shape[1]):
        try:
            vif = variance_inflation_factor(X.values, i)
            vif_values.append(vif if not np.isinf(vif) else 999)  # Treat inf as very high
        except:
            vif_values.append(np.nan)
    
    vif_data["VIF"] = vif_values
    vif_data = vif_data.sort_values('VIF', ascending=False).reset_index(drop=True)
    
    # Filter out protected features from removal consideration
    removable_features = vif_data[~vif_data['feature'].isin(protected_features)]
    
    if len(removable_features) == 0:
        print(f"\nAll remaining features are protected. Process complete.")
        break
    
    max_vif = removable_features['VIF'].max()
    
    if max_vif > max_vif_threshold:
        feature_to_remove = removable_features.iloc[0]['feature']
        print(f"Removing '{feature_to_remove}' (VIF = {max_vif:.2f})")
        X = X.drop(columns=[feature_to_remove])
        df.drop(columns=[feature_to_remove], inplace=True)
    else:
        print(f"\nAll VIF values are below {max_vif_threshold}. Process complete.")
        break

print("\nFinal VIF values:")
print(vif_data)

# Show which protected features remain and their VIF
protected_vif = vif_data[vif_data['feature'].isin(protected_features)]
if not protected_vif.empty:
    print(f"\n✅ Protected features retained:")
    print(protected_vif)
```

*Although 'spend_per_unit' had a high VIF, it was retained as a protected feature due to its business importance from two other features ('annual_spend' and 'condo_size_units') and its potential predictive power for churn. Given the synthetic nature of the dataset, we will keep it for modeling to see if it contributes to performance, while being mindful of its multicollinearity with the other feature 'contract_type_transactional'.*
